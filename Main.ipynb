```python
%run "Operators.ipynb" # imports is_subset, is_less_than
%run "BooleanLA.ipynb" # imports Hasse{Explicit,Implicit}

title = "A cooler Hasse Diagram"
sheet = Sheet()

# implicitly defined relation
linear_example = {
    1: [1, 2, 3, 4, 5, 6, 7]
}
hasse_linear = HasseImplicit("A linear Hasse diagram", linear_example, is_less_than)
sheet.add_diagram(hasse_linear)

# set fun goodness
subset_example = {
    1: [ [1], [1, 2] ],
    2: [ [], [2], [1, 3], [1, 2, 3] ],
    3: [ [3], [2, 3] ]
}
hasse_subset = HasseImplicit("A subset Hasse diagram", subset_example, is_subset)
sheet.add_diagram(hasse_subset)

# explicitly defined relation
disconnected_example = {
    1: ['d'],
    2: ['b'],
    3: ['e'],
    4: ['c'],
    5: ['a'],
    6: ['f']
}
relation = [('a', 'a'), ('a', 'f'), ('b', 'b'), ('b', 'd'),
            ('b', 'e'), ('c', 'a'), ('c', 'c'), ('c', 'f'),
            ('d', 'd'), ('e', 'e'), ('f', 'f')]
hasse_disconnected = HasseExplicit("A disconnected Hasse diagram", disconnected_example, relation)
sheet.add_diagram(hasse_disconnected)

apolygonal_example = {
    1: [2, 3],
    2: [1, 4],
    3: [5],
    4: [6, 7]
}
relation = [(1,2), (1,5), (2, 3), (3, 4), (5, 4), (5, 7), (6, 5)]
hasse_apolygonal = HasseExplicit("An apolygonal Hasse diagram", apolygonal_example, relation)
sheet.add_diagram(hasse_apolygonal)

example_1_1 = {
    1: ['a', 'b'],
    2: ['c'],
    3: ['d']
}
relation = [('a', 'b'), ('b', 'c'), ('c', 'd')]
hasse_e11 = HasseExplicit("Example 1-1", example_1_1, relation)
sheet.add_diagram(hasse_e11)

example_1_2 = {
    1: ['a'],
    2: ['b'],
    3: ['c']
}
relation = [('a', 'b'), ('b', 'c')]
hasse_e12 = HasseExplicit("Example 1-2", example_1_2, relation)
sheet.add_diagram(hasse_e12)

sheet
```
